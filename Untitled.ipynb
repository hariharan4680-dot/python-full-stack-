```python
a = 10
b = 3

print("Arithmetic:")
print(a + b, a - b, a * b, a / b)

print("\nComparison:")
print(a == b, a > b, a < b)

print("\nLogical:")
print(a > 5 and b < 5)

print("\nAssignment:")
a += 2
print(a)

print("\nMembership:")
nums = [1, 2, 3]
print(2 in nums)

print("\nIdentity:")
x = [1]
y = x
print(x is y)
```
