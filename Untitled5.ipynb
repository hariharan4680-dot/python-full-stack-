```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(tips['total_bill'], kde=True)
plt.xlabel("Total Bill")
plt.ylabel("Density")
plt.title("Total Bill Distribution")
plt.show()
```

```python
sns.distplot
sns.distplot(tips['total_bill'], bins = 20)
```

```python
sns.displot(tips['total_bill'], bins=20, kde=True)
plt.show()
```

```python
sns.distplot(tips['total_bill'], hist = False, vertical= True, color = 'red')
```

```python
sns.kdeplot(y=tips['total_bill'], fill=True, color='red')
plt.xlabel("Density")
plt.ylabel("Total Bill")
plt.title("Total Bill Density Plot")
plt.show()
```
