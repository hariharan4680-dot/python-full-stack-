```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv(r"C:\Users\DELL\OneDrive\Documents\forex_data_2020-2025.csv")
df_sample = df.head(20)
bar_width = 0.35
x = range(len(df_sample))
plt.figure(figsize=(12, 6))
plt.bar(x, df_sample['Open'], width=bar_width, label='Open')
plt.bar([i + bar_width for i in x], df_sample['Close'],
        width=bar_width, label='Close')
plt.xlabel("Data Index")
plt.ylabel("Price")
plt.title("Open vs Close Price Comparison")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
labels = ['Open', 'High', 'Low', 'Close']
values = [100, 120, 90, 110]
plt.pie(values, labels=labels, autopct='%1.1f%%')
plt.title("Price Distribution")
plt.show()
```
